```python
import sys
import os

# Adds the parent directory (project root) to the Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from langchain_core.messages import HumanMessage, AIMessage
from src.agent.nodes.supervisor import supervisor_node

# ---------------------------------------------------------
# Test Case 1: Initial Investigation Request
# ---------------------------------------------------------
print("=== Running Test Case 1: New Audit Request ===")

# Mock the state dictionary as if a user just typed a query
mock_state_1 = {
    "messages": [
        HumanMessage(content="Can you please check account C12345 for any transaction discrepancies?")
    ],
    "next_node": "",
    "is_security_breach": False
}

# Invoke the node directly
output_1 = supervisor_node(mock_state_1)

# Verify the output structure
print(f"Test 1 Resulting Payload: {output_1}")
assert output_1["next_node"] == "auditor_worker", "❌ Failed: Should have routed to auditor_worker."
print("✅ Test Case 1 Passed successfully!\n")


# ---------------------------------------------------------
# Test Case 2: Investigation Resolution
# ---------------------------------------------------------
print("=== Running Test Case 2: Wrap Up ===")

# Mock the state as if the worker has already returned data and analyzed it
mock_state_2 = {
    "messages": [
        HumanMessage(content="Can you please check account C12345 for any transaction discrepancies?"),
        AIMessage(content="I am routing this to the data auditor."),
        HumanMessage(content="[System: auditor_worker output returned Risk Profile showing 0 discrepancy]"),
        AIMessage(content="I have analyzed account C12345. The pre-calculated gold profile shows zero balance discrepancies across all historical data, and no system flags are present. The account is clean.")
    ],
    "next_node": "auditor_worker",
    "is_security_breach": False
}

output_2 = supervisor_node(mock_state_2)

print(f"Test 2 Resulting Payload: {output_2}")
assert output_2["next_node"] == "FINISH", "❌ Failed: Should have routed to FINISH."
print("✅ Test Case 2 Passed successfully!")
```
