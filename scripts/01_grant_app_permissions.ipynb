```python
from databricks.sdk import WorkspaceClient
from databricks.sdk.service.workspace import AclPermission

# Initialize the client using your personal credentials
w = WorkspaceClient()

# The Service Principal ID from your error message
app_service_principal_id = "6688de50-45da-4e5d-9b5c-63d8b48fad93"

try:
    # Grant READ permission using the strict Enum object
    w.secrets.put_acl(
        scope="portfolio_secret",
        principal=app_service_principal_id,
        permission=AclPermission.READ  # <-- The fix: Using the Enum instead of a string
    )
    print("✅ Success: Service Principal has been granted READ access to 'portfolio_secret'.")
except Exception as e:
    print(f"❌ Error: {e}")
```
